## Data used is an extract from https://app.datastock.shop/?site_name=Travel%20Listings%20From%20MakeMyTrip

## https://www.makemytrip.com/ is one of the leading travel booking websites in the country. So data is chosen to support and deep dive into the understanding of Indian cutomers travel behaviour. Study is done to explore most visited destinations across the globe as per the available data w.r.t various factor points present below.

## Data description is as follows: 
- Uniq Id- Unique Id of each customer with respect to their package purchase
- Crawl Timestamp- Date and Time of the purchace recorded
- Package Name- Pckage varient and their description
- Page Url- URL of the page from where the particular package was purchased 
- Package Type- Package type depending upon their price and other features
- Company- MakeMyTrip
- Destination- Destinations of the customers
- Itinerary- Details of no of nights spend in each destination
- Places Covered- Places covered in the whole trip
- Travel Date- Date of beginning of the travel
- Hotel 1 Details- Hotel stayed in the first destination
- Hotel 1 Rating- Rating of hotel in the first destinantion in (5points). 
- Hotel 2 Details- Hotel stayed in the second destination
- Hotel 2 Rating- Rating of hotel in the second destinantion in (5points). 
- Hotel 3 Details- Hotel stayed in the third destination
- Hotel 3 Rating- Rating of hotel in the third destinantion in (5points). 
- Start City- Cities from which people travelled, City of departure
- Airline- Airlines used for travel with respect to package destinantions
- Onwards Return Flight Time- Departure and arrival time of the flight
- Price Per Two Persons- Travel Price per two person
- Per Person Price- Travel Price per person
- Sightseeing Places Covered- Siteseeing places visited with respect to their destinantions
- Cancellation Rules- Details of cancellation procedure
